Aim : Use of named entity recognition information extraction technique.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """Apple Inc. is an American multinational technology company headquartered in Cupertino, California.
It was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in 1976.
In 2023, Apple is one of the Big Five American information technology companies, alongside Amazon, Google, Microsoft, and Meta.
"""

tkn = []
doc = nlp(text)

for token in doc:
    tkn.append(token.text)
print(tkn)

for ent in doc.ents:
    print(f"Entity : {ent.text}, Type : {ent.label_}")
```
